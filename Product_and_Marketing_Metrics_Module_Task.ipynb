```python
from IPython.display import display, Markdown

content = """
# Продуктовий аналіз сервісу Booking.com

## 1. Аналіз ключових етапів життя користувача (Activation > Engagement > Retention)

### Activation
**Метрики:**
- Кількість зареєстрованих користувачів
- Частка користувачів, які здійснили перший пошук
- Частка користувачів, які додали житло в обране
- Частка користувачів, які зробили перше бронювання

**Бізнес-зв’язок:**
Чим більше користувачів проходить початковий шлях до бронювання, тим більший шанс на подальші покупки і зростання revenue. Активовані користувачі = потенційні клієнти.

### Engagement
**Метрики:**
- DAU/WAU/MAU (щоденна / тижнева / місячна активність)
- Середня кількість переглядів готелів на одного користувача
- Кількість сесій на одного користувача
- Частота додавання в обране
- Частота фільтрації або сортування пошуку

**Бізнес-зв’язок:**
Високий рівень залученості показує, що користувачі активно шукають і планують поїздки — більше шансів на бронювання і зростання продажів.

### Retention
**Метрики:**
- Repeat Booking Rate (частка користувачів, які зробили повторне бронювання)
- % користувачів, що повернулись протягом 30/60/90 днів
- Середній інтервал між бронюваннями
- NPS у повернених користувачів

**Бізнес-зв’язок:**
Повторні бронювання = стабільний прибуток. Retention критично важливий у висококонкурентному середовищі.
"""

display(Markdown(content))
```

```python
from IPython.display import display, Markdown

colab_content = """
### Монетизаційна модель Booking.com

#### Основна модель монетизації:
Booking.com працює за **моделлю комісії** (Commission-based model). Готелі, апартаменти та інші партнери сплачують Booking.com фіксований відсоток від кожного бронювання (у середньому ~15%).

#### Бізнес-логіка:
Booking.com зацікавлений у **максимізації кількості підтверджених бронювань**, оскільки кожне бронювання = комісія.

---

### Ключові метрики для монетизації

| Метрика | Пояснення |
|--------|-----------|
|  Confirmed Bookings | Підтверджені бронювання напряму впливають на дохід |
|  Revenue per Booking | Сума доходу з одного бронювання |
|  Average Commission Rate | Середній відсоток комісії |
|  Gross Booking Value (GBV) | Загальна сума всіх бронювань |
|  Active Partners | Кількість активних готелів та хостів |
|  Partner Conversion Rate | Частка готелів, що отримали бронювання після публікації |
|  Partner Retention Rate | Повторне використання платформи партнерами |
|  Customer Lifetime Value (CLV) | Довгострокова цінність клієнта |
|  Churn Rate партнерів | Відсоток готелів, які відмовились від співпраці |

---

### Зв’язок із бізнес-цілями:

- Зростання **Confirmed Bookings** напряму збільшує **Revenue**.
- Підтримка високого **Partner Retention Rate** дозволяє зменшити витрати на залучення нових партнерів.
- Аналіз **CLV** клієнтів допомагає ефективно інвестувати у маркетинг.
- Оптимізація **Commission Rate** впливає на привабливість платформи для готелів.

---

 **Висновок**: Booking.com монетизується через комісію з кожного бронювання. Метрики мають вимірювати не лише прибуток, але й здоров’я партнерської екосистеми та ефективність клієнтських рішень.
"""

display(Markdown(colab_content))
```

```python
from IPython.display import display, Markdown

nsm_block = """
## NSM (North Star Metric) для Booking.com

### Чи можливо адаптувати концепцію NSM?

Так!
NSM допомагає фокусувати команду на метриці, яка відображає **довгострокову цінність для користувача** і напряму пов’язана з ростом бізнесу.

---

### Потенційні North Star Metrics для Booking.com

#### 1. **Кількість підтверджених бронювань**
- Відображає реальне використання сервісу.
- Проста, зрозуміла для всіх команд.
- Прямо пов'язана з revenue.

#### 2. **Кількість унікальних клієнтів, які завершили бронювання**
- Фокусується на активних користувачах.
- Показує як acquisition, так і retention.

#### 3. **Total Nights Booked (Загальна кількість заброньованих ночей)**
- Враховує як частоту, так і глибину взаємодії користувачів.
- Добре показує "довжину" користувацької цінності.

#### 4. **Gross Booking Value (GBV)**
- Вимірює фінансовий обсяг бронювань.
- Показує комерційний потенціал продукту.

---

### Рекомендований NSM:
> **Total Confirmed Booked Nights per Month**

 Чому саме ця метрика?
- Поєднує **обсяг** і **частоту** бронювань.
- Балансує між клієнтською активністю та монетизацією.
- Добре масштабується для B2C і B2B.

---

 **Висновок**:
Booking.com може використовувати NSM як спільний фокус для команд Growth, Product, Marketing та Sales. Це дозволить приймати рішення, що впливають саме на цінність для користувача — і прибутковість.
"""

display(Markdown(nsm_block))
```

```python
from IPython.display import display, Markdown

rfm_block = """
## RFM-аналіз користувачів Booking.com

### Що таке RFM?
**RFM (Recency, Frequency, Monetary)** — фреймворк для сегментації користувачів на основі їхньої поведінки:

| Компонент | Пояснення |
|-----------|-----------|
| **Recency** | Коли користувач востаннє зробив бронювання |
| **Frequency** | Як часто користувач бронює на платформі |
| **Monetary** | Скільки грошей витрачає користувач |

---

### Ціль сегментації:
Ідентифікувати найбільш цінних та найменш активних користувачів → персоналізувати маркетинг, активувати “сплячих” або винагороджувати лояльних.

---

### Приклад сегментації (за RFM):

| Сегмент | Опис |
|---------|------|
|  **Champions** | Часто бронюють і нещодавно були активні — основна ціль утримання |
|  **Big Spenders** | Не часто, але з великим чеком — важливі для преміум-пропозицій |
|  **At Risk** | Давно не повертались — важливі для реанімування |
|  **Cold Users** | Майже неактивні — низький пріоритет |
|  **New Customers** | Нещодавно прийшли — потребують onboarding-комунікації |

---

### Рекомендований сегмент для аналізу:
> **“Champions” — користувачі, які здійснили ≥2 бронювань за останні 3 місяці**

#### Метрики для аналізу:
- Кількість бронювань за останні 3 місяці (F)
- Дата останнього бронювання (R)
- Загальна сума витрат (M)
- Частка позитивних відгуків / NPS
- Типи обраних житл (апартаменти, готелі тощо)

---

### Висновок:
RFM дозволяє будувати **персоналізовані кампанії** та **ефективно витрачати маркетинговий бюджет**. Для Booking.com це критично в умовах високої конкуренції.
"""

display(Markdown(rfm_block))
```

```python
from IPython.display import display, Markdown

b2b_block = """
## B2B-складова Booking.com та метрики ефективності

### Що таке B2B у Booking.com?

Booking.com має розвинену B2B-екосистему, де основними партнерами є:
-  Готелі, апартаменти, хостели
-  Транспортні компанії (таксі, прокат авто)
-  Агенти/туроператори через API-платформу
-  SaaS-рішення для керування готелем (extranet, аналітика, автоматизація)

---

### Бізнес-модель B2B:
Booking.com надає партнерам платформу для продажу послуг та бере комісію за підтверджені бронювання. Також пропонує доступ до аналітики, технічної підтримки та маркетингових інструментів.

---

### Метрики для оцінки ефективності B2B-сегменту

| Метрика | Пояснення |
|---------|-----------|
|  **Partner Acquisition Rate** | Кількість нових партнерів за місяць |
|  **Partner Retention Rate** | Скільки партнерів залишаються активними 3+ місяці |
|  **Partner Revenue Contribution** | Частка доходу від B2B-партнерів |
|  **Partner NPS** | Оцінка задоволеності готелів |
|  **Churn Rate партнерів** | Частка готелів, які перестали користуватись платформою |
|  **Listing Quality Score** | Якість контенту партнера (фото, опис, відгуки) |
|  **Support Efficiency** | Середній час вирішення проблем партнера |
|  **Usage of Partner Tools** | Скільки партнерів активно використовують SaaS-рішення Booking |

---

### Висновок:
Ефективність B2B-напрямку напряму впливає на доступність і різноманіття пропозицій для клієнтів. Висока якість сервісу для партнерів — це запорука масштабованого зростання Booking.com.

"""

display(Markdown(b2b_block))
```
